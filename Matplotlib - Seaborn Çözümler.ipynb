```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips
```

#### Soru: Sigara içen ve içmeyen müşterilerin fatura tutarlarının dağılımını gösteren bir histogram oluşturun.

```python
smoker_yes = tips[(tips["smoker"] == "Yes")]["total_bill"]
smoker_no = tips[(tips["smoker"] == "No")]["total_bill"]

plt.hist(smoker_yes, alpha=0.5, label= "Sigara İçenler")
plt.hist(smoker_no, alpha=0.5, label= "Sigara İçmeyenler")
plt.legend()
plt.title("Sigara İçen ve İçmeyenlerin Fatura Dağılımları")
plt.xlabel("Fatura Tutarı")
plt.ylabel("Kişi Sayısı")
```

#### Soru: Veri setindeki "total_bill" değişkeninin dağılımını gösteren bir histogram oluşturun.

```python
plt.hist(tips["total_bill"], label= "Toplam Fatura")
plt.legend()
plt.title("Topla Fatura")
plt.xlabel("Fatura Tutarı")
plt.ylabel("Frekans")
```

#### Soru: Veri setindeki "total_bill" ve "tip" değişkenleri arasındaki ilişkiyi gösteren bir scatterplot oluşturun. Ayrıca, "time" değişkenini kullanarak noktaları renklendirin.

```python
sns.scatterplot(data= tips, x= "total_bill", y= "tip", hue= "time")
```

#### Soru: Veri setindeki "day" ve "total_bill" değişkenlerini kullanarak, her gün için ortalama toplam fatura tutarını gösteren bir barplot oluşturun.

```python
sns.barplot(data= tips, x= "day", y= "total_bill")
```

#### Soru: "tips" veri setinde, her günün bahşiş oranını (yüzde olarak) hesaplayın ve bunları gösteren bir line plot oluşturun.

```python
tips["tip_yüzde"] = (tips["tip"] / tips["total_bill"]) * 100
sns.lineplot(data= tips, x= "day", y= "tip_yüzde")
```

#### Soru: "tips" veri setinde, her gün için fatura tutarının ve bahşiş miktarının ortalamasını hesaplayarak, iki değişken arasındaki ilişkiyi gösteren bir scatterplot oluşturun.

```python
ortalama = tips.groupby("day")[["total_bill","tip"]].mean()

sns.scatterplot(data= ortalama, x= "total_bill", y= "tip", hue= "day")
```

#### Soru: Sigara içen ve içmeyen müşterilerin, farklı günlerde ödedikleri toplam bahşiş miktarını gösteren bir bar plot oluşturun.

```python
sns.barplot(data= tips, x= "smoker", y= "total_bill", hue= "day")
```

#### Soru: Sigara içen ve içmeyen müşterilerin ödedikleri bahşiş miktarlarının dağılımını gösteren bir boxplot oluşturun.

```python
sns.boxplot(data= tips, x= "smoker", y= "tip")
```

#### Soru: Yukarıdan seçeceğiniz 4 farklı plotu, subplot kullanarak çizdirin.

```python
ex, ax = plt.subplots(2, 2)
ax[0, 0].hist(tips["smoker"])
ax[0, 1].scatter(tips["total_bill"], tips["tip"])
ax[1, 1].bar(tips["sex"], tips["tip"])
ax[1, 0].bar(tips["sex"], tips["total_bill"])
```
